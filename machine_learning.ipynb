```python
df = pd.read_csv("https://raw.githubusercontent.com/timothykeating/Final-Project/vedika/demographic_data_NC.csv") 
```

```python
# COMPINE YEAR AND MONTH columns IN ONE column
df['DATE'] = pd.to_datetime(df[['YEAR', 'MONTH']].assign(DAY=1)) 
df
```

```python
df["DATE"].value_counts() 
```

```python
df['month_name'] = df.DATE.dt.month_name() 
```

```python
df_cleanen = df.drop(['YEAR','MONTH','SEX','RACE','MARST','HISPAN','EDUC','COVIDTELEW','DATE'], axis=1) 
```

```python
# Create our features
X = pd.get_dummies(df_cleanen, columns=["gender","education","race","hispanic","marital_status","month_name"])
# Create our target
y = df_cleanen['COVIDUNAW'] 
X
```
