```python
import pandas as pd
```

```python
# Files to load
demographic_data_df = pd.read_csv("cps_00004.csv")
demographic_data_df = demographic_data_df.drop(['HWTFINL','CPSID','ASECFLAG','PERNUM','WTFINL','CPSIDP'], axis=1)
demographic_data_df.head(10)
```

```python
# Filtering data for North Carolina
demographic_data_df_NC = demographic_data_df[demographic_data_df['STATEFIP'] == 37]
demographic_data_df_NC
```

```python
# Filtering target columns to keep valid data and drop 99 values
demographic_data_df_NC = demographic_data_df_NC[demographic_data_df_NC['COVIDUNAW'] != 99]
demographic_data_df_NC.head(10)
```

```python
# Filtering target columns to keep valid data and drop 99 values
demographic_data_df_NC = demographic_data_df_NC[demographic_data_df_NC['COVIDTELEW'] != 99]
demographic_data_df_NC
```

```python
#Dropping State FIP column
demographic_data_df_NC = demographic_data_df_NC.drop(['STATEFIP'], axis=1)
demographic_data_df_NC
```

```python
demographic_data_df_NC['SEX'].value_counts()
```

```python
gender={1:"Male",
     2:"Female"   
}
```

```python
#Applying map function to change categorical data from numbers to labels
demographic_data_df_NC["gender"] = demographic_data_df_NC['SEX'].map(gender)
demographic_data_df_NC.head()
```

```python

demographic_data_df_NC['age'] = pd.cut(x=demographic_data_df_NC['AGE'], bins=[16, 24, 34, 44, 54, 64,90],
                    labels=['16 to 24', '25 to 34', '35 to 44',
                            '45 to 54', '55 to 64','65+'])
```

```python
demographic_data_df_NC
```

```python
#Custom Encoding Months
demographic_data_df_NC['month_le'] = (demographic_data_df_NC["YEAR"]-2020)*12 + demographic_data_df_NC["MONTH"]
demographic_data_df_NC
```

```python
# COMPINE YEAR AND MONTH columns IN ONE column
demographic_data_df_NC['DATE'] = pd.to_datetime(demographic_data_df_NC[['YEAR', 'MONTH']].assign(DAY=1)) 
demographic_data_df_NC
```

```python
#Dropping all Nan values
demographic_data_df_NC.dropna(how='any',inplace = True)
demographic_data_df_NC
```

```python
demographic_data_df_NC = demographic_data_df_NC.drop(['SEX','RACE','MARST','HISPAN','EDUC','COVIDTELEW'], axis=1)
demographic_data_df_NC
```
