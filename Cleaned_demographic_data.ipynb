```python
import pandas as pd
```

```python
# Files to load
demographic_data_df = pd.read_csv("cps_00004.csv")
demographic_data_df = demographic_data_df.drop(['HWTFINL','CPSID','ASECFLAG','PERNUM','WTFINL','CPSIDP'], axis=1)
demographic_data_df.head(10)
```

```python
demographic_data_df_NC = demographic_data_df[demographic_data_df['STATEFIP'] == 37]
demographic_data_df_NC
```

```python
demographic_data_df_NC = demographic_data_df_NC[demographic_data_df_NC['COVIDUNAW'] != 99]
demographic_data_df_NC.head(10)
```

```python
demographic_data_df_NC = demographic_data_df_NC[demographic_data_df_NC['COVIDTELEW'] != 99]
demographic_data_df_NC
```

```python
demographic_data_df_NC = demographic_data_df_NC.drop(['STATEFIP'], axis=1)
demographic_data_df_NC
```

```python
demographic_data_df_NC['SEX'].value_counts()
```
