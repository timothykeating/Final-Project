```python
# COMPINE YEAR AND MONTH columns IN ONE column
df['DATE'] = pd.to_datetime(df[['YEAR', 'MONTH']].assign(DAY=1)) 
df
```

```python
df["DATE"].value_counts() 
```

```python
df_cleanen = df.drop(['YEAR','MONTH','SEX','RACE','MARST','HISPAN','EDUC','COVIDTELEW'], axis=1) 
```

```python
df_cleanen
```

```python
# reorder the columns
df_cleanen = df_cleanen[["DATE","METAREA","COUNTY","AGE","gender","education","race","hispanic","marital_status","COVIDUNAW"]]
df_cleanen
```

```python
# Create our features
X = pd.get_dummies(df_cleanen, columns=["DATE","gender","education","race","hispanic","marital_status"])
# Create our target
y = df_cleanen['COVIDUNAW'] 
X
```
